# üì• Cell 3 ‚Äî LOADING THE DATASET FROM UCI
What it does: Downloads the official UCI ZIP file, locates the CSV file, and reads it correctly with a semicolon separator. Saves an optional local copy.

What you get: A DataFrame (df) with the data already loaded, including the first few rows and the format (rows/columns).
How it contributes: A starting point for the EDA (Enhanced Data Analysis), reproducible and clean. (Uses pandas to explore the file)

```python
uci_zip_url = "https://archive.ics.uci.edu/static/public/697/predict%2Bstudents%2Bdropout%2Band%2Bacademic%2Bsuccess.zip"
zip_bytes = urllib.request.urlopen(uci_zip_url).read()
z = zipfile.ZipFile(io.BytesIO(zip_bytes))


csv_name = [n for n in z.namelist() if n.lower().endswith(".csv")][0]


df = pd.read_csv(
    z.open(csv_name),
    sep=';',
    quotechar='"',
    encoding='utf-8',
    engine='python'
)


os.makedirs("/content/data", exist_ok=True)
df.to_csv("/content/data/data.csv", index=False)

print("Shape:", df.shape)
display(df.head())
```

#üßπ Cell 8 ‚Äî DUPLICATES AND RANGES

What it does: Removes duplicates if present and reports out-of-range values ‚Äã‚Äãin key variables (e.g., Admission grade).

What you get: A dataset without duplicates and a diagnosis of potential range errors.
How it contributes: Performs data cleaning (duplicate removal, plausibility check).

```python

dup_count = df.duplicated().sum()
if dup_count > 0:
    df = df.drop_duplicates().reset_index(drop=True)
print(f"Duplicates removed: {dup_count}")


range_checks = []
if 'Admission grade' in df.columns:
    out_of_range = ~df['Admission grade'].between(0, 200)
    range_checks.append(('Admission grade', int(out_of_range.sum())))
if 'Age at enrollment' in df.columns:
    out_of_range = ~df['Age at enrollment'].between(15, 80)
    range_checks.append(('Age at enrollment', int(out_of_range.sum())))
print("Chequeos de rango:", range_checks)
```

# üß© Cell 9 ‚Äî MISSING ENTRIES AND IMPUTATION


What it does: Counts missing entries per column. If any exist (via parsing), it imputes: numeric ‚Üí median, categorical ‚Üí mode.

What you get: Confirmation of ‚Äúno missing entries‚Äù or, if any exist, an imputed dataset with the declared policy.
How it contributes: Fulfills data imputation (explicit requirement).

```python
na_table = df.isna().sum().sort_values(ascending=False)
display(na_table.head(10))
total_na = int(na_table.sum())

if total_na == 0:
    print("No missing values found. No imputation applied.")
else:
    num_cols = df.select_dtypes(include=[np.number]).columns
    cat_cols = df.select_dtypes(exclude=[np.number]).columns
    for c in num_cols:
        if df[c].isna().any():
            df[c] = df[c].fillna(df[c].median())
    for c in cat_cols:
        if df[c].isna().any():
            df[c] = df[c].fillna(df[c].mode(dropna=True)[0])
    print(f"Charge made. Total previous shortages: {total_na}")
```

#üîó Cell 13 ‚Äî BIVARIATE (Categorical vs. Target)
What it does: Contingency tables and stacked bars of Scholarship holder/Tuition fees up to date versus Target.
What you get: Chart 6 (and 7) + crosstab tables to interpret association.
How it contributes: Bivariate analysis between categorical variables and the class (requirement).

```python
import os
os.makedirs("/content/docs/img", exist_ok=True)

def stacked_bar(df_, col_cat, target='Target', outfile=None):
    ct = pd.crosstab(df_[col_cat], df_[target])
    ct.plot(kind='bar', stacked=True, rot=0)
    plt.title(f'{col_cat} vs {target} (stacked counts)')
    plt.xlabel(col_cat)
    plt.ylabel('Conteo')
    plt.tight_layout()

    if outfile is not None:
        plt.savefig(outfile, dpi=150)

    plt.show()
    return ct

for col in ['Scholarship holder', 'Tuition fees up to date']:
    if col in df.columns and 'Target' in df.columns:
        filename = col.lower().replace(" ", "_").replace("/", "_")
        outfile = f"/content/docs/img/fig_{filename}_vs_target.png"

        ct = stacked_bar(df, col, 'Target', outfile=outfile)
        display(ct.head())
```

#üì¶ Cell 14 ‚Äî BIVARIATE (Numerical vs. Target)

What it does:

For a numerical variable (e.g., Admission grade), it draws a violin for each Target class (Dropout, Graduate, Enrolled).

Each violin shows the full distribution (density) and its median/body, making it more intuitive to see differences in shape (skewness, tails, multimodality).

What you get:

1‚Äì2 violin plots (e.g., for Admission grade and Age at enrollment).

Differences between classes are more easily perceived than with a boxplot full of outliers.

```python
import os
os.makedirs("/content/docs/img", exist_ok=True)

TARGET_COL = 'Target'
NUM_COLS_TO_PLOT = ['Admission grade', 'Age at enrollment']

if TARGET_COL not in df.columns:
    print(f"Column not found '{TARGET_COL}'.")
else:
    classes_present = df[TARGET_COL].dropna().unique().tolist()
    preferred_order = ['Dropout', 'Graduate', 'Enrolled']
    order = [c for c in preferred_order if c in classes_present] + \
            [c for c in classes_present if c not in preferred_order]

    for col in NUM_COLS_TO_PLOT:
        if col in df.columns:
            data = [df.loc[df[TARGET_COL] == cls, col].dropna().values for cls in order]

            if sum(len(x) > 0 for x in data) < 2:
                print(f"Insufficient data for {col}.")
                continue

            plt.figure(figsize=(8, 5))
            plt.violinplot(dataset=data, showmeans=True, showmedians=True, showextrema=True)
            plt.xticks(np.arange(1, len(order) + 1), order)
            plt.title(f'{col} por {TARGET_COL} ‚Äî Violin plot')
            plt.ylabel(col)
            plt.tight_layout()


            safe_col = col.lower().replace(" ", "_").replace("/", "_")
            outfile = f"/content/docs/img/fig_violin_{safe_col}_by_target.png"
            plt.savefig(outfile, dpi=150)

            plt.show()
```

üß™ Celda 15 ‚Äî MULTIVARIADO (Correlaciones)

Tipo: C√≥digo
Qu√© hace: Calcula y dibuja la matriz de correlaci√≥n entre variables num√©ricas.
Qu√© obtienes: Gr√°fico 10, vista global de relaciones lineales entre num√©ricas.
C√≥mo aporta: Cubre el multivariate analysis exigido.

```python
import os
os.makedirs("/content/docs/img", exist_ok=True)

EXCLUIR_SEMESTRES = True
TOP_K = 12


num_cols = df.select_dtypes(include=[np.number]).columns


if EXCLUIR_SEMESTRES:
    mask_1 = ~num_cols.str.contains('Curricular units 1st sem', case=False, regex=False)
    mask_2 = ~num_cols.str.contains('Curricular units 2nd sem', case=False, regex=False)
    num_cols = num_cols[mask_1 & mask_2]

corr_full = df[num_cols].corr()


mean_abs = corr_full.abs().where(~np.eye(len(corr_full), dtype=bool)).mean().sort_values(ascending=False)
top_features = mean_abs.head(min(TOP_K, len(mean_abs))).index
corr_top = corr_full.loc[top_features, top_features]


labels = (corr_top.columns
          .str.replace('Previous qualification (grade)', 'Prev qual (grade)', regex=False)
          .str.replace('Age at enrollment', 'Age enroll', regex=False)
          .str.replace('Tuition fees up to date', 'Fees up-to-date', regex=False)
          .str.replace('Curricular units 1st sem', '1stSem', regex=False)
          .str.replace('Curricular units 2nd sem', '2ndSem', regex=False)
         )

plt.figure(figsize=(10, 8))
plt.imshow(corr_top.values, aspect='equal', vmin=-1, vmax=1)
plt.title(f'Correlation matrix ‚Äî numerical (Top {len(labels)})')
plt.xticks(range(len(labels)), labels, rotation=45, ha='right')
plt.yticks(range(len(labels)), labels)
plt.colorbar()
plt.tight_layout()

# Guardar
plt.savefig("/content/docs/img/fig_corr_top_numeric.png", dpi=150)

# Mostrar
plt.show()
```
